```python
import numpy as np
import pylab as py
import scipy as sp
from scipy import stats
%pylab inline
```

```python
np.random.seed(555)
mymean = 60.
#mystdev=10.
nit = 1000
poi = np.random.poisson(mymean,nit)
py.hist(poi)
py.xlabel("Number of events", fontsize=12)
py.ylabel("Frequency", fontsize=12)
py.title("Poisson Distribution", fontsize=14)
```

```python
myDmean=np.zeros(nit)
for i in range(1000):
    mydistrib=np.random.poisson(mymean,nit)
    myDmean[i]=mydistrib.mean()
```

```python
py.hist(myDmean);
py.xlabel("Number of events", fontsize=12)
py.ylabel("Frequency", fontsize=12)
py.title("Distribution of mean of 1000 Poisson Distributions", fontsize=14)
```

```python
zs=(mymean-myDmean)*np.sqrt(1000) / np.sqrt(mymean)
py.hist(zs);
py.xlabel("Z Statistic", fontsize=12)
py.ylabel("Frequency", fontsize=12)
py.title("Distribution of Z Statistic of 1000 Poisson Distributions", fontsize=14)
```

## fitting the z-statistics with a gaussian model and assessing the goodness of fit with a KS test.
** Null Hypothesis, H0: Distribution of Z Statistics comes from a gaussian distribution, alpha = 0.05 **

```python
sp.stats.kstest(zs, 'norm')
```

## KS: high p-value pvalue>0.05: H0 that sample comes from a Gaussian distribution cannot be rejected

** When the size increases, Poisson distribution can be approximated to a Gaussian Distribution. This is in accordance with Central Limit Theorem **
